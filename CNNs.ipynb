```python
from IPython.display import SVG
from keras.utils import model_to_dot
from keras.models import Sequential
from keras.layers import Dropout, Activation, BatchNormalization, Convolution2D, MaxPooling2D
from keras import metrics
from keras import Input, Model
```

```python
model = Sequential()

model.add(Convolution2D(64,(5,5), padding='same', activation='relu', input_shape=(100,100,1)))
model.add(BatchNormalization())

model.add(Convolution2D(128,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(256,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(128,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(64,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Convolution2D(1,(5,5), padding='same', activation='relu'))

model.summary()

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
inputs = Input(shape=(100,100,1))

x = Convolution2D(128,(5,5), padding='same', activation='relu')(inputs)

dx = Convolution2D(128,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

x = Convolution2D(256,(5,5), padding='same', activation='relu')(x)

dx = Convolution2D(256,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

dx = Convolution2D(256,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

dx = Convolution2D(256,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

dx = Convolution2D(256,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

dx = Convolution2D(256,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

x = Convolution2D(128,(5,5), padding='same', activation='relu')(x)

dx = Convolution2D(128,(5,5), padding='same', activation='relu')(x)
dx = BatchNormalization()(dx)
x = dx + x

outputs = Convolution2D(1,(5,5), padding='same', activation='relu')(x)

model = Model(inputs = inputs, outputs = outputs)

model.summary()

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
SVG(model_to_dot(model).create(prog=’dot’, format=’svg’))
```
