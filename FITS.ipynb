```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from os.path import join
from os import listdir
from glob import glob
import pandas as pd
```

```python
data_dir = 'data/Mesch_QSO2D/STARweak_3XMMJ074405.8+284354/'
```

```python
files = glob(join(data_dir,'*.fits'))
files.sort()
```

```python
hdul = fits.open(files[0])
hdul.info()
hdul[0].header
```

```python
list1 = []
list2 = []
files1 = []
files2 = []
files_ = []
sizes = []
objs = []
exps = []

for file in files:
    obj = fits.getdata(file)
    hdul = fits.open(file)
    exps.append(hdul[0].header['EXPTIME'])
    objs.append(hdul[0].header['OBJECT'])
    sizes.append(obj.shape)
    files_.append(file.split('/')[-1])
    if len(obj.shape) == 2:
        list1.append(obj)
        files1.append(file)
    else:
        list2.append(obj)
        files2.append(file)
        
data = {'File': files_, 'Exptime': exps, 'Object': objs, 'Size':sizes}
    
df = pd.DataFrame(data, columns=['File', 'Object', 'Exptime', 'Size'])
df
```

```python
_, axes = plt.subplots(9, 5, figsize=(20,20))
for i,img in enumerate(list1):
    axes[i//5, i%5].imshow(img)
    axes[i//5, i%5].set_title(files1[i].split('/')[-1])
```

```python
_, axes = plt.subplots(6, 1, figsize=(20,20))
for i,img in enumerate(list2):
    axes[i].plot(img.mean(axis=2))
    axes[i].set_title(files2[i].split('/')[-1])
```

```python
_, axes = plt.subplots(6, 4, figsize=(20,20))
for i,img in enumerate(list2):
    for j in range(img.shape[0]):
        axes[i,j].plot(img[j].mean(axis=0))
        axes[i,j].set_title(files2[i].split('/')[-1])
```
